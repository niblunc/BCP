```python
from scipy import stats
import numpy as np
np.random.seed(12345678)
```

```python
rvs1 = stats.norm.rvs(loc=5,scale=10,size=625)
rvs2 = stats.norm.rvs(loc=5,scale=10,size=625)
```

```python
x=np.reshape(np.array(rvs1),(25,25))
```

```python
y=np.reshape(np.array(rvs2),(25,25))
```

```python
y.shape
```

```python
y
```

```python
stats.ttest_rel(x,y)
```
