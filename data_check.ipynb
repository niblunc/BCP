```python
import glob
import os
import numpy 
import pandas as pd
from pandas.api.types import is_string_dtype
from pandas.api.types import is_numeric_dtype
from functools import reduce
```

```python
basepath='/Users/gracer/Google Drive/BCP/data'
```

```python
file_dict = {}
for item in glob.glob(os.path.join(basepath,'*.csv')):
    key=item.split('/')[-1].split('.csv')[0]
    file_dict.update({key:item})
```

```python
data_dict={}
for key, value in file_dict.items():
    print(key)
    data_dict[key]=pd.read_csv(value, sep=",", encoding='latin1')
```

```python
import pdb
```

```python
summary_dict={}
for key, value in data_dict.items():
    print(key)
    summary_dict[key]={}
    IDcols=value.filter(regex="ID")
    main_list = list(set(value.columns)-set(IDcols.columns))
    summary_dict[key]['summary']={}
    for col in IDcols.columns:
        if col == 'UserID':
            print('skip')
        else:
            summary_dict[key]['summary']['counts_%s'%col]=IDcols[col].value_counts()
            summary_dict[key]['summary']['unique_%s'%col]='There are %s %s unique IDs'%(IDcols[col].unique().shape, col)
    for col in value[main_list]:
        print(col)
        if is_string_dtype(value[col]):
            print('this is the column %s and it is a string'%col)
            summary_dict[key]['NA_%s'%col]=[value[col].isna().sum(),'this is a string type, nas' ]
            summary_dict[key]['Not_NA_%s'%col]=[value[col].notna().sum(),'this number of not na']
        else:
            print('this is the column %s and it is NOT a string'%col)
            summary_dict[key][col]=[value[col].isna().sum(),value[col].describe()]
```

```python
for key, value in summary_dict.items():
    print(key)
    for k,v in summary_dict[key]['summary'].items():
        if 'unique' in k:
            print(v)
```
